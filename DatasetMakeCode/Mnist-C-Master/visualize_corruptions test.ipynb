```python
import os
import random
from torchvision.datasets import MNIST


train_mnist = MNIST("./data/", train=True, download=True)
test_mnist = MNIST("./data/", train=False, download=True)


selected_labels = [2]
count_per_label = 1250  
total_images = len(selected_labels) * count_per_label

save_directory_all = "B"



os.makedirs(save_directory_all, exist_ok=True)


selected_indices = []
count_per_label_dict = {label: 0 for label in selected_labels}

for i in range(len(test_mnist)):
    _, true_label = test_mnist[i]
    if true_label in selected_labels and count_per_label_dict[true_label] < count_per_label:
        selected_indices.append(i)
        count_per_label_dict[true_label] += 1


while sum(count_per_label_dict.values()) < total_images:
    for i in range(len(test_mnist)):
        _, true_label = test_mnist[i]
        if true_label in selected_labels and count_per_label_dict[true_label] < count_per_label:
            selected_indices.append(i)
            count_per_label_dict[true_label] += 1

            if sum(count_per_label_dict.values()) == total_images:
                break


random.seed(42) 
random.shuffle(selected_indices)


IMAGES = []
LABELS = []

for index, i in enumerate(selected_indices):
    image, true_label = test_mnist[i]
    IMAGES.append(image)
    LABELS.append(true_label)


for i, (image, label) in enumerate(zip(IMAGES, LABELS)):
    new_filename_all = f"image_{i + 1}.png"
    new_filepath_all = os.path.join(save_directory_all, new_filename_all)
    image.save(new_filepath_all)
```

```python
import os

def delete_images_with_pattern(folder_path, pattern, start_index, end_index):
    for i in range(start_index, end_index + 1):
        file_path = os.path.join(folder_path, f"{pattern}{i}.png")
        if os.path.exists(file_path):
            os.remove(file_path)
            print(f"Deleted: {file_path}")

folder_path = "B"  
pattern = "image_"  
start_index = 1  
end_index = 1000 
delete_images_with_pattern('B', pattern, start_index, end_index)
```

```python
import os
import random

def shuffle_and_rename_images(folder_path):

    image_files = [f for f in os.listdir(folder_path) if f.endswith('.png') or f.endswith('.jpg')]
    
    random.shuffle(image_files)
    
    for i, filename in enumerate(image_files):
        original_filepath = os.path.join(folder_path, filename)
        new_filename = f'image_{i + 1}.png' 
        new_filepath = os.path.join(folder_path, new_filename)
        os.rename(original_filepath, new_filepath)
        print(f"Renamed: {filename} to {new_filename}")

folder_path = "A"  
shuffle_and_rename_images(folder_path)
```
